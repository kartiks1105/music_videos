```python
from bs4 import BeautifulSoup
from youtubesearchpython import VideosSearch
import pandas as pd
```

```python
file_path = '/Users/sriv.kartik/Desktop/Development/MusicVideos Project/EVERYTHING.txt'
df = pd.read_csv(file_path, sep='\t')

df.head(2)
```

```python
df.columns
```

```python
cols_to_drop = ['Location','My Rating', 'Last Skipped', 'Last Played', 'Comments', 'Equalizer', 'Kind', 'Volume Adjustment',
               'Sample Rate', 'Bit Rate', 'Date Modified', 'Track Count', 'Track Number', 'Disc Count', 'Disc Number',
               'Time', 'Size', 'Movement Name', 'Movement Count', 'Movement Number', 'Work', 'Grouping']

df = df.drop(columns = cols_to_drop)
```

```python
df.head()
```

```python
def search_music_video(song_name):
    videos = VideosSearch(f"{song_name} official music video", limit=1)
    results = videos.result()
        
    if 'result' in results and results['result']:
        link = results['result'][0]['link']
        return link
    else:
        return None

df['music_video_link'] = df['Name'].apply(search_music_video)
```
